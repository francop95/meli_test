En una primera instancia, se busca encontrar cuales son los experimentos que lograron mayor cantidad de compras. Por lo que de esta dataframe, me quedare solo con los resultados BUY

```python
df_buy=df[df['event_name']=='BUY']
```

```python
df_buy
```

Ahora, de estos productos que se compraron, para una misma persona, se buscara todo los experimentos asociados a ese item_id.

```python
df_filt=df[df['user_id'].isin(df_buy['user_id']) & df['item_id'].isin(df_buy['item_id'])]
df_filt
```

Con estos filtros nos aseguramos de tener todos los experimentos desde el momento en que la persona ve un producto (item_id) hasta el momento en el que lo compra ('event_name' == 'BUY')

```python
df_filt['experiments']=df_filt['experiments'].map(lambda x: x.lstrip('{').rstrip('}'))
df_resultados=pd.DataFrame(df_filt['experiments'].str.split(",\s+").apply(lambda x: pd.Series(x).value_counts()).sum().sort_values(ascending=False)).reset_index().rename(columns={"index":"experiment",0:"count"})
```

```python
df_resultados['id_variante']=df_resultados['experiment'].str.split('=').str[1]
df_resultados['experiment']=df_resultados['experiment'].str.split('=').str[0]
df_resultados=df_resultados[['experiment','id_variante','count']]
df_resultados.sort_values(by=['experiment'], ascending=False)
```

Con esto, tengo la lista de todos los experimentos y sus variant id que se realizaron a una persona X, que termino realizando una compra de un producto Y, pasando por todos los event_name.
